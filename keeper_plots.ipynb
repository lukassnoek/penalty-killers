```python
import nibabel as nib
import numpy as np
import nilearn.plotting as npl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tvals_pos = nib.load('RESULTS/TRAIN/WITHIN_SUBS/pos_tvals_clustercorr.nii.gz')
affine = tvals_pos.affine
tvals_pos = tvals_pos.get_data()
tvals_neg = nib.load('RESULTS/TRAIN/WITHIN_SUBS/neg_tvals_clustercorr.nii.gz').get_data()
tvals = tvals_pos + tvals_neg * -1
tvals = nib.Nifti1Image(tvals, affine=affine)
npl.plot_stat_map(tvals, draw_cross=False, threshold=1.7, display_mode='x', cut_coords=8)
```

```python
npl.plot_stat_map(tvals, draw_cross=False, threshold=1.7, display_mode='y', cut_coords=8)
```
